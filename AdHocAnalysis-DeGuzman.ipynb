```python
import pandas as pd
```

```python
data = pd.read_json("transaction-data-adhoc-analysis.json")
data
```

```python
itemListed = data.assign(transaction_items = data['transaction_items'].str.split(';'))
```

```python
singleItemOnly = itemListed[itemListed['transaction_items'].apply(len) == 1][['transaction_items','transaction_value']]
```

```python
singleItemOnly = singleItemOnly.assign(
    itemCount = lambda x: x['transaction_items'].apply(lambda x: int(x[0].split(',')[-1].strip('()x'))),
    transaction_items = lambda x: x['transaction_items'].apply(lambda x: '{} - {}'.format(x[0].split(',')[0],x[0].split(',')[1]))
  )
```

```python
singleItemOnly = singleItemOnly.drop_duplicates(subset="transaction_items", keep="first")
singleItemOnly
```

```python
singleItemOnly = singleItemOnly.assign(unit_value = singleItemOnly['transaction_value']/singleItemOnly['itemCount'])
```

```python
singleItemOnly.index = singleItemOnly['transaction_items']
```

```python
singleItemOnly.drop(columns=['transaction_items','transaction_value','itemCount'],inplace=True)
singleItemOnly
```

```python
data = itemListed.explode('transaction_items').reset_index(drop=True)
data
```

```python
data = data.assign(transaction_date = pd.to_datetime(data["transaction_date"]))

data = data.assign(
    item_count = lambda x: x['transaction_items'].apply(lambda x: int(x.split(',')[-1].strip('()x'))),
    transaction_items = lambda x: x['transaction_items'].apply(lambda x: '{} - {}'.format(x.split(',')[0],x.split(',')[1]))
  )

data['transaction_value'] = data['item_count'] * data['transaction_items'].apply(lambda x: singleItemOnly.loc[x].values[0])

data
```

# Count of each item sold per month

```python
data.groupby([data['transaction_items'],data['transaction_date'].dt.year,data['transaction_date'].dt.month])['item_count'].sum()
```

#Total sale value per item per month

```python
data.groupby([data['transaction_items'],data['transaction_date'].dt.year,data['transaction_date'].dt.month])['transaction_value'].sum()
```

# Repeaters, Inactive, Engaged

```python
userTable = data[['name','transaction_date']]
userTable = userTable.assign(transaction_date = userTable['transaction_date'].apply(lambda x: (x.year,x.month)))
userTable
```

```python
userTable = pd.pivot_table(userTable, index=['transaction_date'], columns=['name'], aggfunc=any, fill_value=False)
userTable
```

```python
repeaters = userTable.apply(
    lambda x: [False if i==0 else (True if x[i-1] and x[i] else False) for i,j in enumerate(x)],
  ).transpose().sum()
inactive = userTable.apply(
    lambda x: [False if i==0 else ((True if x[i]==False else False) if any(x[:i]) else False) for i,j in enumerate(x)]
  ).transpose().sum()
engaged = userTable.apply(
    lambda x: [True if all(x[:i]) else False for i,j in enumerate(x)]
  ).transpose().sum()

pd.DataFrame({'repeaters':repeaters,'inactive':inactive,'engaged':engaged}).transpose()
```
